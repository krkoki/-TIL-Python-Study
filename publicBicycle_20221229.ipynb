# 일별 이용 현황

## 대여날짜 컬럼 추가
* date

```python
df['대여날짜'] = df['대여일시'].dt.date
```

```python
df.head()
```

## 대여날짜 별 대여 건수
* groupby
* to_frame

```python
# 대여날짜 별 대여건수 추출
borrow = df.groupby('대여날짜')['대여일시'].count()
borrow.columns = ['대여건수']
borrow
```

```python
# 대여날짜 별 대여건수 시각화
plt.plot(borrow.index, borrow.values)
plt.show()
```

## 대여날짜 별 이용시간
* groupby
* to_frame

```python
# 대여날짜 별 이용시간 추출
usetime = df.groupby('대여날짜')['이용시간'].count()
usetime
```

```python
# 대여날짜 별 이용시간 시각화
plt.plot(usetime.index, usetime.values)
plt.show()
```

## 대여날짜 별 이용 거리

```python
# 대여날짜 별 이용거리 추출
distance = df.groupby('대여날짜')['이용거리'].count()
distance
```

```python
# 대여날짜 별 이용거리 시각화
plt.plot(distance.index, distance.values)
plt.show()
```

## 데이터프레임 합치기

```python
df_add = pd.concat([borrow, usetime, distance], axis=1)
df_add
```

# 시간대별 대여/반납 현황

## 대여시간, 반납시간 컬럼 추가

```python
df['대여시간'] = df['대여일시'].dt.hour
```

```python
df['반납시간'] = df['반납일시'].dt.hour
```

## 시간대별 대여/반납 현황

```python
# 시간대별 대여현황
time_borrow = df['대여시간'].value_counts()
time_borrow
```

```python
# 시간대별 반납현황
time_return = df['대여시간'].value_counts()
time_return
```

## 시각화

```python
# 시간대별 공공자전거 대여건수 시각화
plt.bar(time_borrow.index, time_borrow.values)
plt.xlabel('대여시간')
plt.ylabel('대여건수')
plt.show()
```

```python
# 시간대별 공공자전거 반납건수 시각화
plt.bar(time_return.index, time_return.values)
plt.xlabel('대여시간')
plt.ylabel('반납건수')
plt.show()
```
