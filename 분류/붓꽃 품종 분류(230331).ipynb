```python
# 붓꽃의 품좀을 머신러닝을 이용하여 꽃잎과 꽃받침의 크기를 기반으로 분류
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from matplotlib import font_manager, rc
#한글 처리를 위해 폰트 설정
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

```python
# csv 파일을 로딩
df = pd.read_csv("../data/iris/iris.csv")
print(df.head())
```

```python
# 품종을 기준으로 한 평균값들
df.groupby("Species").mean()
```

```python
# 파이차트
sums_by_species = df.groupby('Species').sum()
var = 'SepalWidth'
sums_by_species[var].plot(kind='pie', fontsize=20)
plt.ylabel(var, horizontalalignment='left')
#plt.savefig('iris_pie.png')
```

```python
# 품종별 변수 합계
sums_by_species = df.groupby('Species').sum()
print(sums_by_species)
var = 'SepalWidth'

# rot 글자회전(rotation)
# 꽃받침 너비(cm) 기준 분류 그래프
sums_by_species[var].plot(kind='bar', fontsize=15, rot=30)
```

```python
# 품정별 꽃잎 길이
for spec in df['Species'].unique():
    subdf = df[df['Species'] == spec]
    subdf['PetalLength'].plot(kind='hist', alpha=0.4, label=spec)
    plt.legend(loc='upper right')
```

```python
df['Species'].value_counts()
```

```python
import seaborn as sns
sns.countplot('Species', data=df)
plt.title('분류\n (0: setosa, 1: versicolor, 2: virginica)', fontsize=14)
```

```python
cols = ["SepalLength", 'SepalWidth', 'PetalLength', 'PetalWidth']
X = df[cols]
y = df['Species']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=0)
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
print("학습용:", model.score(X_train, y_train))
print("검증용:", model.score(X_test, y_test))
```

```python
# confusion matrix(혼동행렬, 오분류표) 출력
from sklearn.metrics import confusion_matrix
pred = model.predict(X_test)
cm = confusion_matrix(y_test, pred)
cm
```

```python
df_cm = pd.DataFrame(cm, index=[0, 1, 2],
                     columns=[0, 1, 2])
df_cm
```

```python
import seaborn as sns
from matplotlib import pyplot as pyplot
%matplotlib inline
sns.set(font_scale=1.4)
plt.figure(figsize=(9, 6))
plt.title("Confusion Matrix")
sns.heatmap(df_cm, annot=True)
# 생상이 밝으면 높은 숫자, 색상이 어두우면 낮은 숫자
```
