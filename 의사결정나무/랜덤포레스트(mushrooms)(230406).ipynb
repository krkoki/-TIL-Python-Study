```python
# 의사결정나무의 단점 : 과거의 데이터는 잘 맞추지만 새로운 데이터에 대한 예측력이 약함
# 과적합화를 방지할 수 있는 대표적인 방법 중 하나가 랜덤포레스트
# 여러개의 의사결정나무를 만들고 투표를 통해 다수결로 결과를 결정함
# 처리가 빠르고 분류 정밀도가 높다.
import pandas as pd
# 데이터 로딩
# 첫번째 필드 : 독버섯 p, 식용 e
df = pd.read_csv("../data/mushrooms/mushrooms.csv")
df.head()
```

```python
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder
# 카테고리형 데이터를 숫자형으로 변환하는 함수
# 주의할 부분이 있음 (가중치가 붙을 수 있음)
lbl = LabelEncoder()
for col in df.columns:
    df[col]=lbl.fit_transform(df[col])
    print(lbl.classes_)
```

```python
X=df.iloc[:, 1:]
y=df['class']
df['class'].value_counts() 
# 0 식용 4208, 1 독버섯 3916
```

```python
# 언더샘플링
from imblearn.under_sampling import RandomUnderSampler
X_sample, y_sample = RandomUnderSampler(random_state=0).fit_resample(X, y)
X_samp = pd.DataFrame(data=X_sample, columns = X.columns)

# 원핫인코딩
X_samp = pd.get_dummies(X_samp, columns=X_samp.columns, drop_first=True)
y_samp = pd.DataFrame(data=y_sample, columns=['class'])
df_samp = pd.concat([X_samp, y_samp], axis= 1)
df_samp.head()
```

```python
df_samp['class'].value_counts() 
# 0 식용 3916, 1 독버섯 3916
```

```python
X = df_samp.iloc[:, :-1]
y = df_samp['class']
print(X.head())
print(y.head())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
# 모형 평가
from sklearn.metrics import accuracy_score, classification_report
pred = model.predict(X_test)
score = accuracy_score(y_test, pred)
report = classification_report(y_test, pred)
print("정답률 : ", score)
print("리포트 : \n", report)
# 100% 분류 완료
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
def plot_tree(model):
    n_features = X.shape[1] # 독립변수의 개수
                    #[샘플수, 변수개수]
    plt.rcParams["figure.figsize"]=(16,90)
    plt.barh(range(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features), df_samp.iloc[:, :-1])
    plt.xlabel("importances")
    plt.ylabel("variables")
    plt.ylim(-1, n_features)
```

```python
plot_tree(model)
```
