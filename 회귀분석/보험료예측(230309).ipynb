```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('d:/workspace/Python2/data/insurance/insurance.csv')
data.head()
```

```python
data2 = pd.read_csv('d:/workspace/Python2/data/insurance/insurance2.csv')
data2.head()
```

```python
# 다중 회귀분석

# 명목형 변수는 제외하고 연속형 변수만 독립변수로 선택
# 독립변수 : 연속형 변수만 사용하고 범주형 변수는 사용하지 않음
df2 = data2.iloc[:, [0,2,3]]
print(df2.head())
```

```python
# 종속변수
y2 = data2.iloc[:, [9]]
print(y2.head())
```

```python
import statsmodels.api as sm

model = sm.OLS(y2, df2)
result = model.fit()

print("회귀계수\n", result.params)
print("R-squared", result.rsquared)
print("P-value", result.pvalues)
```

```python
# 요약 결과 출력
print(result.summary())
```

```python
# 유의하지 않은 children 변수 제거
df2 = data2.iloc[:, [0,2]]
print(df2.head())

#종속변수
y2 = data2.iloc[:, [9]]
print(y2.head())

model = sm.OLS(y2, df2)
result = model.fit()

print("회귀계수\n", result.params)
print("R-squared", result.rsquared)
print("P-value", result.pvalues)

#요약 결과 출력
print(result.summary())
```
