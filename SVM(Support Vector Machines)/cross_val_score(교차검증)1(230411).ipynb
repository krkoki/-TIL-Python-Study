```python
import pandas as pd
df = pd.read_csv('../data/customer/customer.csv')
df.head()
```

```python
X = df[df.columns[:-1]]
y = df['Churn']
y.value_counts()
```

```python
# 언더샘플링
from imblearn.under_sampling import RandomUnderSampler
X_sample, y_sample = RandomUnderSampler(random_state=0).fit_resample(X, y)
X_samp = pd.DataFrame(data=X_sample, columns=df.columns[:-1])
y_samp = pd.DataFrame(data=y_sample, columns=['Churn'])
df_samp = pd.concat([X_samp, y_samp], axis=1)
df_samp['Churn'].value_counts()
```

```python
X = df_samp[df_samp.columns[:-1]]
y = df_samp['Churn']
```

```python
import matplotlib.pyplot as plt
X.boxplot()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X_scaled = pd.DataFrame(scaler.transform(X))
```

```python
%matplotlib inline
X_scaled.boxplot()
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import BaggingClassifier
svm = SVC()
n_estimators = 10
model = BaggingClassifier(base_estimator=svm, # 기본 분류 모형
                          n_estimators=n_estimators, # 분류기수
                          max_samples=1/n_estimators) # 
```

```python
import numpy as np
from sklearn.model_selection import cross_val_score
costs = np.logspace(-10, 3, 14) # 10의 -10승부터 10의 3승까지 14개
scores = []
for c in costs:
    model.base_estimator.C = c
    # 교차검증
    this_scores = cross_val_score(model, X_scaled, y, cv=3)
    m = np.mean(this_scores)
    print(c, '==>', m)
    scores.append(m)
print('max:', max(scores))
print('idx:', np.argmax(scores))
print('C:', costs[np.argmax(scores)])
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (16, 5)
plt.semilogx(costs, scores)
locs, labels = plt.yticks()
plt.yticks(locs)
plt.ylabel('CV score')
plt.xlabel('Parameter C')
plt.ylim(0.4, 1)
plt.show()
```
