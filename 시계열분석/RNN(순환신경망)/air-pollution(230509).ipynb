```python
import pandas as pd
df = pd.read_csv('../../data/time/pollution.csv', index_col=0)
df
```

```python
df.boxplot()
```

```python
from sklearn.preprocessing import LabelEncoder
import numpy as np
values = df.values
encoder = LabelEncoder()

# w_dr 변수 정수인코딩
values[:, 4] = encoder.fit_transform(values[:, 4])
values = values.astype('float32')
np.unique(values[:,4])
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled = scaler.fit_transform(values)
pd.DataFrame(scaled).boxplot()
plt.show()
```

```python
# 쉬프트
def shift(train):
    timestamp = 4
    X = []
    y = []
    for i in range(timestamp, len(train)):
        X.append(train[i-timestamp:i, :]) # pollution ~ rain
        y.append(train[i, 0:1]) # pollution
    return X, y
```

```python
X, y = shift(scaled)
X = np.array(X)
y = np.array(y)
print(X.shape)
print(y.shape)
```

```python
# 학습용:검증용, 8:2(4년:1년)
n_train = int(len(X)*.2)+1
X_train, X_test = X[n_train:, ], X[:n_train, ]
print(X_train.shape, X_test.shape)

y_train, y_test = y[n_train:, ], y[:n_train, ]
print(y_train.shape, y_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM
model = Sequential()
# return_sequences = False 기본값, 마지막 스텝만 내보내는 옵션
# return_sequences = True, 모든 스텝에 내보내는 옵션(LSTM을 여러층으로 만들 경우 사용)
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50))
model.add(Dropout(0.2))
model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mse')
model.summary()
```

```python
model.fit(X_train, y_train, validation_split=0.2, epochs=10, batch_size=32)
```

```python
from sklearn.metrics import mean_squared_error
pred = model.predict(X_test)
mse = mean_squared_error(y_test, pred)
rmse = np.sqrt(mse)
rmse
```

```python
d = scaled[:n_train, :]
y_predicted = np.concatenate((pred, d[:n_train, 1:]), axis=1)
y_tested = np.concatenate((y_test, d[:n_train, 1:]), axis=1)
```

```python
y_predicted = scaler.inverse_transform(y_predicted)
y_tested = scaler.inverse_transform(y_tested)
y_predicted = y_predicted[:,0:1]
y_tested = y_tested[:,0:1]
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_tested, y_predicted)
rmse = np.sqrt(mse)
rmse
```

```python
plt.figure(figsize=(10,6))
plt.plot(y_predicted[:100,:], color= 'g',label = 'Predict')
plt.plot(y_tested[:100,:] , color = 'r',label = 'Real')
plt.xlabel("Date")
plt.ylabel("Pollution")
plt.legend()
plt.show()
```
