# 소방청 타지역 검색

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from tqdm.notebook import tqdm
```

```python
url = 'https://www.nfa.go.kr/nfa/introduce/status/firestationidfo/?searchDistance=10&searchMode=distance&myX=37.5605672&myY=126.9433486&searchKeyword='
driver = webdriver.Chrome('chromedriver')
driver.get(url)
time.sleep(2)
```

```python
driver.find_element(By.ID,'searchKeyword').send_keys('광주')
driver.find_element(By.ID,'fsSearchBtn').click() # id = # 안붙여도됨.
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
sta = soup.select('.stations-list > li')
```

```python
st = sta[0]
st
```

```python
# 관서명
name = st.select_one('.title').get_text()
# 주소
addr = st.select_one('address').get_text()
# 전화번호
tel = st.select_one('.tel').get_text()
```

```python
print(name, addr, tel)
```

```python
lines = []
for st in sta:
    name = st.select_one('.title').get_text()
    addr = st.select_one('address').get_text()
    tel = st.select_one('.tel').get_text()
    lines.append([name, addr, tel])
df = pd.DataFrame(lines, columns = ['관서명', '주소', '전화번호'])
```
