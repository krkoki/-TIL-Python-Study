운동장에서 키 순서대로 서기

키가 제일 작은 학생을 맨 앞에 세우고

남은 명의 학생 중 제일 작은 학생을 두번째에 세우고

```python
# 쉽게 설명한 선택 정렬
# : a 입력 리스트
# : 출력 정렬된 새 리스트
# 주어진 리스트에서 최소값의 위치를 돌려주는 함수

def find_min_idx(a):
    n = len(a)
    min_idx = 0
    for i in range(1, n):
        if a[i] < a[min_idx]:
            min_idx = i
    return min_idx

def sel_sort(a):
    result = [] # 새 리스트를 만들어 정렬된 값을 저장
    while a: # 주어진 리스트에 값이 남아있는 동안 계속
        min_idx = find_min_idx(a) # 리스트에 남아 있는 값중 최소값의 위치
        value = a.pop(min_idx) # 찾은 최소값을 빼내어 value에 저장
        result.append(value) # value를 결과 리스트 끝에 추가
    return result

d = [2, 4, 5, 1, 3, 6, 9, 10, 7, 8]
print(sel_sort(d))
```

```python
# 선택 정렬(일반적인 코드)
# 입력 : 리스트 a
# 출력 : 없음(입력으로 주어진 a가 정렬됨)

def sel_sort(a):
    n = len(a)
    for i in range(0, n - 1):
        min_idx = i
        for j in range(i + 1, n):
            if a[j] < a[min_idx]:
                min_idx = j
            print(i, j, a)
        a[i], a[min_idx] = a[min_idx], a[i] # 두 값을 바꾸기
        print(i, j, a)

d = [2, 4, 5, 1, 3, 6, 9, 10, 7, 8]
sel_sort(d)
print(d)
```

```python
# 내림차순 선택 정렬(일반적인 코드)
# 입력 : 리스트 a
# 출력 : 없음(입력으로 주어진 a가 정렬됨)

def sel_sort(a):
    n = len(a)
    for i in range(0, n - 1):
        max_idx = i # 최소값 대신 최대값을 찾아야함
        for j in range(i + 1, n):
            if a[j] > a[max_idx]: # 부등호 방향 뒤집기
                max_idx = j
            print(i, j, a)
        a[i], a[max_idx] = a[max_idx], a[i]
        print(i, j, a)

d = [2, 4, 5, 1, 3, 6, 9, 10, 7, 8]
sel_sort(d)
print(d)
```
