# 지니차트 크롤링

```python
import requests
from bs4 import BeautifulSoup
url = 'https://mw.genie.co.kr/chart/'
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'}
req = requests.get(url, headers=header)
soup = BeautifulSoup(req.text, 'html.parser')
```

```python
soup.find_all('ul')
```

```python
top = soup.select_one('body > section > ul')
```

```python
#순위 
rank = int(top.select_one('.ranking.text-center em').get_text())
```

```python
rank
```

```python
#제목
title = top.select_one('.track.ellipsis').get_text().strip()
```

```python
title
```

```python
#가수
artist = top.select_one('.artist.text-gray.ellipsis').get_text()
```

```python
artist
```

```python
line = [] 
for tr in top:
    try:
        rank = int(tr.select_one('.ranking.text-center em').get_text())
    except:
        rank = 0
    try:   
        title = tr.select_one('.track.ellipsis').get_text().strip()
    except:
        pass
    try:
        artist = tr.select_one('.artist.text-gray.ellipsis').get_text()
    except:
        pass
    line.append([rank, title, artist])
```

```python
import pandas as pd
df = pd.DataFrame(line, columns=['순위', '제목', '가수'])
df
```
